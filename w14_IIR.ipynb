## Contoh Desain IIR dengan Metode Pole Zero Placement

Spesifikasi Filter
1. HPF
2. Fs = 24kHz
3. Fc = 12kHz

```python
# Import library
from scipy import signal
import matplotlib.pyplot as plt

# definisi zero
b = [1, -2, 1]
# definisi pole
a = [1, -1, 0.34]
# definisi pengali k
k = 0.6

# Cek frekuensi response
[w, h] = signal.freqz(b,a,fs=24000)

# Plot frekuensi rensponse
plt.plot(w, k*abs(h))
```

### Contoh Implementasi LPF

Spesifikasi filter
1. LPF
2. Fs = 10kHz
3. Fc = 1kHz
4. Transition = 500kHz
5. Karena transition 500kHz, maka, zero ada di 1.5kHz

```python
# Import library
from scipy import signal
import matplotlib.pyplot as plt

b = [1, -1.1756, 1]
a = [1, -1.2944, 0.64]

[w, h] = signal.freqz(b,a, fs=10000)

plt.figure(1)
plt.plot(w, abs(h))

# Plot dengan pencerminan
[w2, h2] = signal.freqz(b,a, fs=10000, whole=True)
plt.figure(2)
plt.plot(w2, abs(h2))
```

```python
# Import library
from scipy import signal
import matplotlib.pyplot as plt

b = [1, -1.1756, 1]
a = [1, -1.2944, 0.64]
k = 1/2.4 # kalikan dengan faktor k karena gain lebih dari 1

[w, h] = signal.freqz(b,a, fs=10000)

plt.figure(1)
plt.plot(w, k*abs(h)) # kalikan dengan k

# Plot dengan pencerminan
[w2, h2] = signal.freqz(b,a, fs=10000, whole=True)
plt.figure(2)
plt.plot(w2, k*abs(h2)) # kalikan dengan k
```
