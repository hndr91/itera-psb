```python
# fir1 pada Matlab ekivalen dengan firwin di scipy
# import library
from scipy import signal, pi
import numpy as np
import matplotlib.pyplot as plt

# Definisikan spek filter
n = 53 # panjang filter -> orde + 1
fc_ = 1250 # freq cut-off terkoreksi
hfs = 4000 # setengah fs -> fs kita 8000Hz

# Kalkulasi h(n)
b = signal.firwin(n, fc_/hfs, window='hamming', pass_zero='highpass')
np.set_printoptions(precision=4) # bulatkan 4 angkat koma
np.set_printoptions(suppress=True) # hilangkan nilai e
print(b)

# Plot h(n)
plt.figure(1)
plt.plot(b)

# Kalkulasi FFT dari h(n)
w, h = signal.freqz(b, 1, whole=True)
plt.figure(2)
plt.plot(w/pi, abs(h))
```
