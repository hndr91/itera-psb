## Soal 1

Buatlah sebuah sinyal sinus dummy dengan kententuan,
- Kombinasi frekuensi 2, 30, 400 Hz
- Amplitudonya adalah 0.5, 1 dan 0.1
- Sampling rate 5kHz
- Durasi 2 detik

## Jawab Soal 1

```python
import numpy as np
import matplotlib.pyplot as plt

# sampling rate
sr = 5000.0
# sampling interval
ts = 1.0/sr
# durasi
t = np.arange(0,2,ts)

# sinyal 2Hz
freq = 2
amp = 0.5  
x = amp * np.sin(2*np.pi*freq*t)

# sinyal 30Hz
freq2 = 30
amp2 = 1   
x2 = amp2 * np.sin(2*np.pi*freq2*t)

# sinyal 400Hz
freq3 = 400
amp3 = 0.1
x3 = amp3 * np.sin(2*np.pi*freq3*t)

# sinyal noise --> 5Hz + 30Hz + 400Hz
noise = x + x2 + x3

# plot gambar
plt.figure(figsize = (20, 15))

plt.subplot(411, title='Sinyal 2Hz')
plt.plot(t, x, 'b')
plt.ylabel('Amplitude')

plt.subplot(412, title='Sinyal 30Hz')
plt.plot(t, x2, 'b')
plt.ylabel('Amplitude')

plt.subplot(413, title='Sinyal 400Hz')
plt.plot(t, x3, 'b')
plt.ylabel('Amplitude')

plt.subplot(414, title='Sinyal Noise')
plt.plot(t, noise, 'b')
plt.ylabel('Amplitude')

plt.xlabel('Time (s)')
plt.show()
```

## Soal 2

Lakukan BPF dengan Butterworth dengan ketentuan,
- Orde 3
- Rentang bandpass 10 – 100 Hz

```python
# Lowpass Filter dengan Buttworth orde 3

from scipy import signal
bi, ai = signal.butter(3, [10,100], fs=5000, btype='bandpass') # di return dalam bentuk polynomial

yi_butfilt = signal.lfilter(bi, ai, noise)

plt.figure(figsize=(20,3))
plt.plot(t, yi_butfilt, 'b')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
plt.title('Hasil filter')
plt.show()
```
